1. What is a parameter?
- A parameter is a value (like a weight or coefficient) that a machine learning model learns during training to map inputs to outputs.

2. What is correlation? What does negative correlation mean?
- Correlation measures the strength and direction of a linear relationship between two variables.
- Negative correlation means as one variable increases, the other decreases.

3. Define Machine Learning. What are the main components in Machine Learning?
- Machine Learning is a method where computers learn patterns from data without being explicitly programmed.
- Main components: data, features, model, loss function, optimizer, evaluation metrics.

4. How does loss value help in determining whether the model is good or not?
- Loss value shows how far model predictions are from actual values. Lower loss means better model performance.

5. What are continuous and categorical variables?
- Continuous: numeric with infinite values (e.g. height, age).
- Categorical: represent categories/labels (e.g. color, gender).

6. How do we handle categorical variables in Machine Learning?
- By converting them into numerical form so models can process them.

7. What are the common techniques? What do you mean by training and testing a dataset?
- Common techniques: Label Encoding, One-Hot Encoding, Target Encoding.
- Training dataset: used to build the model.
- Testing dataset: used to evaluate model’s performance on unseen data.

8. What is sklearn.preprocessing?
- A module in scikit-learn that provides tools for data scaling, encoding, normalization, and transformation.

9. What is a Test set?
- A portion of data set aside to evaluate model performance after training.

16. What is causation? Explain difference between correlation and causation with an example.
- Causation: one variable directly affects another.
- Correlation: variables move together but one doesn’t cause the other.
- Example: Ice cream sales and drowning deaths correlate due to heat, but ice cream doesn’t cause drowning.

17. What is an Optimizer? What are different types of optimizers? Explain each with an example.
- Optimizer updates model parameters to reduce loss.
- Types:

Gradient Descent: adjusts parameters by average gradient

SGD (Stochastic Gradient Descent): updates on small batches

Adam: adaptive learning rate
Example:

21. What is feature scaling? How does it help in Machine Learning?
- Adjusts features to a common scale (e.g., 0-1 or mean 0 std 1).
- Helps models converge faster and prevents domination by large-scale features.
